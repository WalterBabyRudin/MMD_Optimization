```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2021)

def get_sparsity(sol_hist):
    N, d = np.shape(sol_hist)
    A = np.sum(sol_hist,1)
    
    Sparsity_hist = np.concatenate((A,np.array([1])), axis=0)
    
    Sparsity_hist = Sparsity_hist/d
    x_hist = np.arange(N+1) + 1
    return Sparsity_hist, x_hist
    

    
    
sol_hist_40 = np.load("sol_hist_d_40.npy")
Sparsity_hist_40, x_hist = get_sparsity(sol_hist_40)

plt.rcParams["font.family"] = "arial"
#plt.style.use("ggplot")
fig, axes = plt.subplots(figsize=(6,6))
axes.plot(x_hist, Sparsity_hist_40, label=r'MMD-NTK', 
            linestyle = '-', marker = 'h', color='red',ms=2,linewidth=1)

axes.set_xlim([-1,50])
axes.set_ylim([0,1.05])
axes.grid(True)
axes.tick_params(labelsize=18)
axes.set_xlabel('Iteration Number', fontsize=15)
axes.set_ylabel(r'Sparsity Rate', fontsize=15)
axes.set_title(r'Performance for $d=40$',fontsize=15)

plt.savefig('plt_d_40.pdf',bbox_inches='tight')
```
